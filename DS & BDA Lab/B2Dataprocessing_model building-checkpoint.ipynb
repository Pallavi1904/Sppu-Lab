```python
# Error Correcting
```

```python
def remove_outliers(column):
    p = column.quantile(0.25)
    q = column.quantile(0.75)
    r = q-p
    threshold = 1.5 * r
    outlier = (column < (p - threshold)) | (column > (p + threshold))
    return column[~outlier]
```

```python
df.head()
```

```python
col_name = ['so2', 'spm', 'rspm']
for col in col_name:
    df[col] = remove_outliers(df[col])
```
